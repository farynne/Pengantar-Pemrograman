```python
import pandas as pd                           # import pandas untuk membaca dan mengolah data
from scipy.stats.mstats import gmean, hmean  # import geometric mean & harmonic mean
import matplotlib.pyplot as plt              # import matplotlib untuk grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  
# membaca file Excel dari sheet 'Supp_traces'

a_mean = my_dataset.Rb.mean()  
# menghitung arithmetic mean (rata-rata biasa)

g_mean = gmean(my_dataset['Rb'])  
# menghitung geometric mean (rata-rata geometri, cocok untuk data log-normal)

h_mean = hmean(my_dataset['Rb'])  
# menghitung harmonic mean (sensitif terhadap nilai kecil)

print('-------')
print('arithmetic mean')
print("{0:.1f} [ppm]".format(a_mean))  
# menampilkan arithmetic mean

print('-------')
print('geometric mean')
print("{0:.1f} [ppm]".format(g_mean))  
# menampilkan geometric mean

print('-------')
print('harmonic mean')
print("{0:.1f} [ppm]".format(h_mean))  
# menampilkan harmonic mean
print('-------')

fig, ax = plt.subplots()  
# membuat figure dan axes

ax.hist(my_dataset.Lu, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8)  
# membuat histogram data Lu

ax.axvline(a_mean, color='blue', label='Arithmetic mean', linewidth=4)  
# garis untuk mean biasa

ax.axvline(g_mean, color='yellow', label='Geometric mean', linewidth=4)  
# garis untuk geometric mean

ax.axvline(h_mean, color='pink', label='Harmonic mean', linewidth=4)  
# garis untuk harmonic mean

ax.set_xlabel('Lu [ppm]')  
# label sumbu X

ax.set_ylabel('Probability density')  
# label sumbu Y

ax.legend()  
# menampilkan legenda
```

```python
import pandas as pd              # import pandas untuk membaca dan mengolah data
import matplotlib.pyplot as plt # import matplotlib untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  
# membaca file Excel dari sheet 'Supp_traces'

variance = my_dataset['Rb'].var()  
# menghitung varians data Rb (ukuran sebaran, satuan ppm^2)

stddev = my_dataset['Rb'].std()  
# menghitung standar deviasi (akar varians, satuan ppm)

print('-------')
print('Variance')
print("{0:.0f} [square ppm]".format(variance))  
# menampilkan varians tanpa desimal

print('-------')
print('Standard Deviation')
print("{0:.0f} [ppm]".format(stddev))  
# menampilkan standar deviasi tanpa desimal
print('-------')

fig, ax = plt.subplots()  
# membuat figure dan axes

ax.hist(my_dataset.Rb, bins=20, density=True, edgecolor='k', label='Measurements Hist')  
# membuat histogram data Rb (dalam bentuk densitas)

ax.axvline(my_dataset['Rb'].mean() - stddev, color='red', label=r'mean - 1$\sigma$', linewidth=4)  
# garis vertikal untuk batas bawah (mean - 1 standar deviasi)

ax.axvline(my_dataset['Rb'].mean() + stddev, color='yellow', label=r'mean + 1$\sigma$', linewidth=4)  
# garis vertikal untuk batas atas (mean + 1 standar deviasi)

ax.axvspan(my_dataset['Rb'].mean() - stddev, my_dataset['Rb'].mean() + stddev,
           alpha=0.1, color='pink', label=r'mean $\pm$ 1$\sigma$')  
# area berwarna yang menunjukkan rentang mean ± 1σ

ax.set_xlabel('Rb [ppm]')  
# label sumbu X

ax.set_ylabel('Probability density')  
# label sumbu Y

ax.legend()  
# menampilkan legenda
```

```python
import pandas as pd              # import pandas untuk membaca dan mengolah data
import numpy as np               # import numpy untuk perhitungan numerik (persentil)
import matplotlib.pyplot as plt # import matplotlib untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  
# membaca file Excel dari sheet 'Supp_traces'

q1 = np.percentile(my_dataset.Rb, 25, method='midpoint')  
# menghitung kuartil pertama (Q1 / persentil ke-25)

q3 = np.percentile(my_dataset.Rb, 75, method='midpoint')  
# menghitung kuartil ketiga (Q3 / persentil ke-75)
  
iqr = q3 - q1  
# menghitung Interquartile Range (IQR = Q3 - Q1)

print('-------')
print('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr))  
# menampilkan nilai IQR tanpa desimal
print('-------')

fig, ax = plt.subplots()  
# membuat figure dan axes

ax.hist(my_dataset.Rb, bins='auto', density=True, edgecolor='k', label='Measurements Hist')  
# membuat histogram data Rb

ax.axvline(q1, color='brown', label='Q1', linewidth=4)  
# garis vertikal untuk Q1

ax.axvline(q3, color='red', label='Q3', linewidth=4)  
# garis vertikal untuk Q3

ax.axvspan(q1, q3, alpha=0.1, color='lime', label='Interquaritle range (IQR)')  
# area berwarna yang menunjukkan rentang IQR (50% data tengah)

ax.set_xlabel('Rb [ppm]')  
# label sumbu X

ax.set_ylabel('Probability density')  
# label sumbu Y

ax.legend()  
# menampilkan legenda
```

```python
import pandas as pd              # import pandas untuk membaca dan mengolah data
import numpy as np               # import numpy untuk perhitungan numerik (histogram, dll)
import matplotlib.pyplot as plt # import matplotlib untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  
# membaca file Excel dari sheet 'Supp_traces'

a_mean = my_dataset.Nb.mean()  
# menghitung mean (rata-rata aritmatika) dari data Nb

median = my_dataset.Nb.median()  
# menghitung median (nilai tengah setelah data diurutkan)

hist, bin_edges = np.histogram(my_dataset['Nb'], bins=20, density=True)  
# membuat histogram untuk mengetahui distribusi data

modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1]) / 2  
# menghitung modus (nilai paling sering muncul) dari histogram

fig, ax = plt.subplots()  
# membuat figure dan axes

ax.hist(my_dataset.Nb, bins=20, density=True, edgecolor='k', label="Measurements Hist")  
# membuat histogram data Nb

ax.axvline(modal_value, color='red', label='Modal Value', linewidth=4)  
# garis vertikal untuk modus

ax.axvline(median, color='gold', label='Median Value', linewidth=4)  
# garis vertikal untuk median

ax.axvline(a_mean, color='purple', label='Arithmetic mean', linewidth=4)  
# garis vertikal untuk mean

ax.set_xlabel('Nb [ppm]')  
# label sumbu X

ax.set_ylabel('Probability density')  
# label sumbu Y

ax.legend()  
# menampilkan legenda
```

```python
import numpy as np                      # import numpy untuk perhitungan numerik
from scipy.stats import skew           # import fungsi skew untuk menghitung kemencengan (skewness)

a_mean = my_dataset.Rb.mean()          # menghitung mean (rata-rata) dari data Rb
median = my_dataset.Rb.median()        # menghitung median (nilai tengah)

hist, bin_edges = np.histogram(my_dataset['Rb'], bins=20, density=True)  
# membuat histogram untuk menentukan distribusi data

modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1]) / 2  
# menghitung modus dari histogram (nilai dengan frekuensi tertinggi)

standard_deviation = my_dataset['Rb'].std()  
# menghitung standar deviasi (ukuran penyebaran data)

a1 = (a_mean - modal_value) / standard_deviation  
# Pearson’s first coefficient of skewness (berdasarkan mean dan modus)

a2 = 3 * (a_mean - median) / standard_deviation  
# Pearson’s second coefficient of skewness (berdasarkan mean dan median)

g1 = skew(my_dataset['Rb'])  
# Fisher-Pearson coefficient of skewness (menggunakan fungsi bawaan scipy)

print('-------')  
# garis pembatas output

print("Pearson's first coefficient of skewness: {:.2f}".format(a1))  
# menampilkan nilai skewness versi Pearson 1

print("Pearson's 2nd moment of skewness: {:.2f}".format(a2))  
# menampilkan nilai skewness versi Pearson 2

print("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))  
# menampilkan skewness versi Fisher (yang paling umum dipakai)

print('-------')  
# garis pembatas output
```
